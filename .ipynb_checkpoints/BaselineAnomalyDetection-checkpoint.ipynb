```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Sample data: user login times (in seconds)
login_times = np.array([30, 32, 31, 29, 35, 33, 30, 28, 31, 34, 30, 29, 32, 31, 30])

# Calculate mean and standard deviation
mean = np.mean(login_times)
std_dev = np.std(login_times)

# Define thresholds
lower_threshold = mean - 2 * std_dev
upper_threshold = mean + 2 * std_dev

# Identify anomalies
anomalies = login_times[(login_times < lower_threshold) | (login_times > upper_threshold)]

# Plotting
plt.hist(login_times, bins=10, alpha=0.7, color='blue')
plt.axvline(lower_threshold, color='red', linestyle='dashed', linewidth=1)
plt.axvline(upper_threshold, color='red', linestyle='dashed', linewidth=1)
plt.title('User Login Times with Anomaly Thresholds')
plt.xlabel('Login Time (seconds)')
plt.ylabel('Frequency')
plt.show()

print("Anomalies detected:", anomalies)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Sample data: user login times (in seconds)
login_times = np.array([30, 32, 31, 29, 35, 33, 30, 28, 31, 34, 30, 29, 32, 31, 30])

# Calculate the mean (average) of the login times
mean = np.mean(login_times)

# Calculate the standard deviation of the login times
std_dev = np.std(login_times)

# Define thresholds for anomaly detection
# Lower threshold is set to mean - 2 standard deviations
lower_threshold = mean - 2 * std_dev

# Upper threshold is set to mean + 2 standard deviations
upper_threshold = mean + 2 * std_dev

# Identify anomalies: login times that fall outside the defined thresholds
anomalies = login_times[(login_times < lower_threshold) | (login_times > upper_threshold)]

# Plotting the histogram of login times
plt.hist(login_times, bins=10, alpha=0.7, color='blue')

# Add a dashed red line for the lower threshold
plt.axvline(lower_threshold, color='red', linestyle='dashed', linewidth=1)

# Add a dashed red line for the upper threshold
plt.axvline(upper_threshold, color='red', linestyle='dashed', linewidth=1)

# Set the title of the plot
plt.title('User Login Times with Anomaly Thresholds')

# Label the x-axis
plt.xlabel('Login Time (seconds)')

# Label the y-axis
plt.ylabel('Frequency')

# Display the plot
plt.show()

# Print the detected anomalies
print("Anomalies detected:", anomalies)
```
