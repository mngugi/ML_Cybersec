```python
import matplotlib.pyplot as plt

# Define impact categories and their corresponding colors
impact_categories = ["Minor", "Serious", "Critical", "Catastrophic"]
colors = ["green", "yellow", "orange", "red"]

# Create a bar chart to visualize categories with colors
plt.figure(figsize=(8, 4))
plt.bar(impact_categories, [1, 1, 1, 1], color=colors, width=0.6)

# Add labels and title
plt.title("Impact Categories and Their Colors", fontsize=14)
plt.xlabel("Impact Category", fontsize=12)
plt.ylabel("Severity Level (Illustrative)", fontsize=12)

# Customize tick marks and remove y-axis ticks for aesthetics
plt.xticks(fontsize=10)
plt.yticks([])

# Show the chart
plt.tight_layout()
plt.show()
```
