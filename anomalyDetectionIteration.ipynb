```python
import numpy as np
import matplotlib.pyplot as plt

# Quadratic iteration parameters
a, b, c = 1, -2, 1  # Example coefficients
iterations = 100
x_values = np.zeros(iterations)

# Initial value
x_values[0] = 0.5

# Perform quadratic iterations
for n in range(1, iterations):
    x_values[n] = a * x_values[n-1]**2 + b * x_values[n-1] + c

# Calculate mean and standard deviation for anomaly detection
mean = np.mean(x_values)
std_dev = np.std(x_values)

# Define thresholds
lower_threshold = mean - 2 * std_dev
upper_threshold = mean + 2 * std_dev

# Identify anomalies
anomalies = x_values[(x_values < lower_threshold) | (x_values > upper_threshold)]

# Plotting
plt.plot(x_values, label='Quadratic Iteration Results')
plt.axhline(lower_threshold, color='red', linestyle='dashed', linewidth=1, label='Lower Threshold')
plt.axhline(upper_threshold, color='red', linestyle='dashed', linewidth=1, label='Upper Threshold')

# Ensure x
```
