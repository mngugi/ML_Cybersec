```python
import numpy as np
import matplotlib.pyplot as plt
import pyfiglet # type: ignore

def display_title(title):
    ascii_art = pyfiglet.figlet_format(title)
    print(ascii_art)

if __name__ == "__main__":
    display_title("Anomaly Detection Using Quadratic Iteration")

def iterate_equation(a, b, c, x0, num_iterations):
    """Iteratively compute values of x using the equation x_{n+1} = ax_n^2 + bx_n + c."""
    x_values = [x0]
    for _ in range(num_iterations):
        x_next = a * x_values[-1]**2 + b * x_values[-1] + c
        x_values.append(x_next)
    return x_values

# Parameters for the equation
a = 1.5
b = -0.3
c = 0.1
x0 = 0.1  # Initial value
num_iterations = 50  # Number of iterations

# Compute the iterations
x_values = iterate_equation(a, b, c, x0, num_iterations)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(range(num_iterations + 1), x_values, marker='o', label='Infection Proportion')
plt.title("Dynamics of Malware Spread", fontsize=16)
plt.xlabel("Iteration (Time Step)", fontsize=14)
plt.ylabel("Proportion Infected", fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```
