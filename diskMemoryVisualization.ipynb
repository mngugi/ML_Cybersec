```python
import psutil
import time

def display_disk_usage():
    while True:
        # Get disk usage statistics
        disk_usage = psutil.disk_usage('/')
        
        # Clear the console (works on Windows and Unix)
        print("\033c", end="")  # For Unix
        # print("\033[H\033[J", end="")  # For Windows

        # Display disk usage
        print("Disk Usage Information:")
        print(f"Total: {disk_usage.total / (1024 ** 3):.2f} GB")
        print(f"Used: {disk_usage.used / (1024 ** 3):.2f} GB")
        print(f"Free: {disk_usage.free / (1024 ** 3):.2f} GB")
        print(f"Percentage Used: {disk_usage.percent}%")
        
        # Wait for a second before updating
        time.sleep(1)

if __name__ == "__main__":
    display_disk_usage()
```
