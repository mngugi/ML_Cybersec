### Graph Weight Analysis for the Education Triad (Students, Parents, Teachers) Using Python

1. Introduction
`A Graph Weight Analysis` can help model the interactions between Students (S), Parents (P), and Teachers (T). This will allow us to analyze collaboration effectiveness, identify weak links, and optimize communication strategies.

2. Graph Representation
- Nodes (Vertices):

- S → Student
- P → Parent
- T → Teacher
- Edges (Connections):

- S ↔ P (Student-Parent communication)
- S ↔ T (Student-Teacher feedback)
- P ↔ T (Parent-Teacher collaboration)
- Edge Weights (Interaction Strength):

- High weight → Frequent and effective interaction.
- Low weight → Rare or ineffective interaction.

3. Implementing Graph Weight Analysis Using Python (networkx)
We will use the networkx library to build and analyze the Education Triad graph.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed weighted graph
G = nx.DiGraph()

# Define nodes (Education Triad)
nodes = ["Student", "Parent", "Teacher"]
G.add_nodes_from(nodes)

# Define edges with weights (interaction strengths)
edges = [
    ("Student", "Parent", 3),  # 3 = Medium interaction (e.g., occasional check-ins)
    ("Student", "Teacher", 5),  # 5 = Strong interaction (e.g., daily class interactions)
    ("Parent", "Teacher", 2),  # 2 = Weak interaction (e.g., rare meetings)
]

# Add edges with weights
G.add_weighted_edges_from(edges)

# Draw the graph with edge labels
plt.figure(figsize=(5, 5))
pos = nx.circular_layout(G)  # Position nodes in a circular layout
nx.draw(G, pos, with_labels=True, node_color="skyblue", edge_color="gray", node_size=3000, font_size=12)
edge_labels = {(u, v): d for u, v, d in G.edges(data="weight")}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=12)

plt.title("Education Triad - Graph Weight Analysis")
plt.show()
```

4. Understanding the Graph Output
The output is a weighted directed graph, where:

- Nodes represent Students, Parents, and Teachers.
- Edges represent interactions between them.
- Edge weights indicate the strength of collaboration.
- Higher weight (5) = Strong communication (e.g., Student-Teacher).
- Lower weight (2) = Weak communication (e.g., Parent-Teacher).


5. Advanced Analysis
We can extend this model by:

- Calculating Centrality (Who has the most influence?)

```python
centrality = nx.degree_centrality(G)
print("Centrality Scores:", centrality)
```

6. Finding the Weakest Link (Which connection needs improvement?)


### Find lowest weight

```python
weakest_link = min(edges, key=lambda x: x[2])  
print("Weakest connection:", weakest_link)
```

```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Number of students
num_students = 100

# Generate unique IDs
students = [f"S{i}" for i in range(1, num_students + 1)]
parents = [f"P{i}" for i in range(1, num_students + 1)]
teachers = [f"T{(i % 10) + 1}" for i in range(1, num_students + 1)]  # 10 teachers shared

# Generate engagement scores (scale: 1-10)
S_P_engagement = np.random.randint(1, 10, num_students)  # Student-Parent
S_T_feedback = np.random.randint(1, 10, num_students)  # Student-Teacher
P_T_meetings = np.random.randint(1, 10, num_students)  # Parent-Teacher

# Compute performance score
performance_labels = ["Poor", "Average", "Good"]
performance_scores = S_P_engagement * 0.3 + S_T_feedback * 0.5 + P_T_meetings * 0.2
performance_categories = pd.cut(performance_scores, bins=[0, 4, 7, 10], labels=performance_labels)

# Create DataFrame
df = pd.DataFrame({
    "Student ID": students,
    "Parent ID": parents,
    "Teacher ID": teachers,
    "S ↔ P (engagement)": S_P_engagement,
    "S ↔ T (feedback)": S_T_feedback,
    "P ↔ T (meetings)": P_T_meetings,
    "Student Performance": performance_categories
})

# Save dataset to CSV
csv_filename = "education_triad_data.csv"
df.to_csv(csv_filename, index=False)

print(f"Dataset Created and Saved as '{csv_filename}'")
#print("Dataset Created and Saved as 'education_triad_data.csv'")
#print(df.head())  # Show first 5 rows
df.head()
```
