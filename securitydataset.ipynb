```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Ignore all warnings
warnings.filterwarnings("ignore")


# Set plot style
sns.set(style="whitegrid")

# List of numerical columns
numerical_columns = [
    'URL_LENGTH', 'NUMBER_SPECIAL_CHARACTERS', 'CONTENT_LENGTH',
    'WHOIS_REG_YEAR', 'UPDATE_YEAR', 'TCP_CONVERSATION_EXCHANGE',
    'DIST_REMOTE_TCP_PORT', 'REMOTE_IPS', 'APP_BYTES', 'UDP_PACKETS',
    'SOURCE_APP_PACKETS', 'REMOTE_APP_PACKETS', 'SOURCE_APP_BYTES',
    'REMOTE_APP_BYTES', 'APP_PACKETS', 'DNS_QUERY_TIMES'
]



# Plot histograms for numerical columns
fig, axes = plt.subplots(nrows=8, ncols=2, figsize=(15, 40))
fig.suptitle('Distribution of Numerical Columns', fontsize=20)

for ax, col in zip(axes.flatten(), numerical_columns):
    sns.histplot(dt[col], bins=30, kde=True, ax=ax)
    ax.set_title(col)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()


```

**The histograms provide a clear view of the distribution of numerical columns:**

1.URL_LENGTH: Mostly concentrated around 40-60, with a few URLs being significantly longer.
2.NUMBER_SPECIAL_CHARACTERS: Most URLs have 6-15 special characters.
3.CONTENT_LENGTH: Highly skewed, with many small values and a few very large values.
4.WHOIS_REG_YEAR: Shows a uniform distribution with some peaks, especially around the year 2000.
5.UPDATE_YEAR: More recent years are more frequent, indicating recent updates.
6.TCP_CONVERSATION_EXCHANGE: Majority of values are below 500.
7.DIST_REMOTE_TCP_PORT: Most common values are around 2-5.
8.REMOTE_IPS: Generally low, with many entries having fewer than 5 remote IPs.
9.APP_BYTES: Highly skewed, with a few very large values.
10.UDP_PACKETS: All values are zero, indicating no UDP packets in the dataset.
11.SOURCE_APP_PACKETS: Mostly concentrated below 50, with some higher values.
12.REMOTE_APP_PACKETS: Similar to SOURCE_APP_PACKETS, concentrated below 50.
13.SOURCE_APP_BYTES: Highly skewed with many small values.
14.REMOTE_APP_BYTES: Similar to SOURCE_APP_BYTES, highly skewed.
15.APP_PACKETS: Most values are below 50.
16.DNS_QUERY_TIMES: Majority of values are between 0-5

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Ignore all warnings
warnings.filterwarnings("ignore")
# List of categorical columns
categorical_columns = [
    'CHARSET', 'SERVER', 'CACHE_CONTROL', 'WHOIS_COUNTRY',
    'WHOIS_STATE_CITY', 'DOMAIN_NAME', 'TIPO'
]

# Plot bar charts for categorical columns
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 20))
fig.suptitle('Distribution of Categorical Columns', fontsize=20)

for ax, col in zip(axes.flatten(), categorical_columns):
    if col in dt.columns:
        sns.countplot(y=dt[col], order=dt[col].value_counts().index, ax=ax)
        ax.set_title(col)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
